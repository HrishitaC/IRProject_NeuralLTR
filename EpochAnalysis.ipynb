```python
def extractaccuracyperepoch(data1):
    ndcg5perepoch1 = []
    ndcg10perepoch1 = []
    ndcg10trainperepoch = []
    ndcg30perepoch1 = []
    ndcg60perepoch1 = []
    mrr5perepoch1 = []
    mrr10perepoch1 = []
    mrr10trainperepoch = []
    mrr30perepoch1 = []
    mrr60perepoch1 = []
    for i in range(100):
        epochstring = 'Epoch : '+str(i)
        startindex = data1.find(epochstring)
        ndcg5ind = data1.find('Val ndcg_5 ',startindex)
        ndcg10ind = data1.find('Val ndcg_10 ',startindex)
        ndcg10trainind = data1.find('Train ndcg_10 ',startindex)
        ndcg30ind = data1.find('Val ndcg_30 ',startindex)
        ndcg60ind = data1.find('Val ndcg_60 ',startindex)
        mrr5ind = data1.find('Val mrr_5 ',startindex)
        mrr10ind = data1.find('Val mrr_10 ',startindex)
        mrr10trainind = data1.find('Train mrr_10 ',startindex)
        mrr30ind = data1.find('Val mrr_30 ',startindex)
        mrr60ind = data1.find('Val mrr_60 ',startindex)
        ndcg5 = data1[ndcg5ind+len('Val ndcg_5 '):ndcg5ind+len('Val ndcg_5 ')+10]
        ndcg10 = data1[ndcg10ind+len('Val ndcg_10 '):ndcg10ind+len('Val ndcg_10 ')+10]
        ndcg10train = data1[ndcg10trainind+len('Train ndcg_10 '):ndcg10trainind+len('Train ndcg_10 ')+10]
        ndcg30 = data1[ndcg30ind+len('Val ndcg_30 '):ndcg30ind+len('Val ndcg_30 ')+10]
        ndcg60 = data1[ndcg60ind+len('Val ndcg_60 '):ndcg60ind+len('Val ndcg_60 ')+10]
        mrr5 = data1[mrr5ind+len('Val mrr_5 '):mrr5ind+len('Val mrr_5 ')+10]
        mrr10 = data1[mrr10ind+len('Val mrr_10 '):mrr10ind+len('Val mrr_10 ')+10]
        mrr10train = data1[mrr10trainind+len('Train mrr_10 '):mrr10trainind+len('Train mrr_10 ')+5]
        mrr30 = data1[mrr30ind+len('Val mrr_30 '):mrr30ind+len('Val mrr_30 ')+10]
        mrr60 = data1[mrr60ind+len('Val mrr_60 '):mrr60ind+len('Val mrr_60 ')+10]
        ndcg5perepoch1.append(float(ndcg5))
        ndcg10perepoch1.append(float(ndcg10))
        ndcg10trainperepoch.append(float(ndcg10train))
        ndcg30perepoch1.append(float(ndcg30))
        ndcg60perepoch1.append(float(ndcg60))
        mrr5perepoch1.append(float(mrr5))
        mrr10perepoch1.append(float(mrr10))
        mrr10trainperepoch.append(float(mrr10train))
        mrr30perepoch1.append(float(mrr30))
        mrr60perepoch1.append(float(mrr60))
    return ndcg5perepoch1,ndcg10perepoch1,ndcg10trainperepoch,ndcg30perepoch1,ndcg60perepoch1,mrr5perepoch1,mrr10perepoch1,mrr10trainperepoch,mrr30perepoch1,mrr60perepoch1
```

```python
#validation accuracies per epoch
ndcg5perepoch1,ndcg10perepoch1,ndcg10trainperepoch1,ndcg30perepoch1,ndcg60perepoch1,mrr5perepoch1,mrr10perepoch1,mrr10trainperepoch1,mrr30perepoch1,mrr60perepoch1 = extractaccuracyperepoch(data1)
ndcg5perepoch2,ndcg10perepoch2,ndcg10trainperepoch2,ndcg30perepoch2,ndcg60perepoch2,mrr5perepoch2,mrr10perepoch2,mrr10trainperepoch2,mrr30perepoch2,mrr60perepoch2 = extractaccuracyperepoch(data2)
ndcg5perepoch3,ndcg10perepoch3,ndcg10trainperepoch3,ndcg30perepoch3,ndcg60perepoch3,mrr5perepoch3,mrr10perepoch3,mrr10trainperepoch3,mrr30perepoch3,mrr60perepoch3 = extractaccuracyperepoch(data3)
ndcg5perepoch4,ndcg10perepoch4,ndcg10trainperepoch4,ndcg30perepoch4,ndcg60perepoch4,mrr5perepoch4,mrr10perepoch4,mrr10trainperepoch4,mrr30perepoch4,mrr60perepoch4 = extractaccuracyperepoch(data4)
ndcg5perepoch5,ndcg10perepoch5,ndcg10trainperepoch5,ndcg30perepoch5,ndcg60perepoch5,mrr5perepoch5,mrr10perepoch5,mrr10trainperepoch5,mrr30perepoch5,mrr60perepoch5 = extractaccuracyperepoch(data5)
```

```python
qids = set()
for line in lines:
    startindex = data.find('qid:')
    endindex = data.find(' 1:',startindex)
    qid = line[startindex+4:endindex]
    qids.add(qid)
print(len(qids))
sampledqids25 = random.sample(qids,int(0.25*len(qids)))
sampledqids50 = random.sample(qids,int(0.5*len(qids)))
sampledqids75 = random.sample(qids,int(0.75*len(qids)))
```
